```python
import pandas as pd
```

```python
csv_file = "USAEduLevel.csv"
# csv_file = "the file_nameXXXX.csv"
USAEduLevel = pd.read_csv(csv_file)
```

```python
print(type(USAEduLevel))
USAEduLevel.head(50)
```

```python
import matplotlib.pyplot as plt
```

```python

USAEduLevel_UnitedStates = USAEduLevel[USAEduLevel['country'] == 'United States']

USAEduLevel_UnitedStates.plot(kind='line',
                     x='year',
                     y='Percentage%',
                     title='UnitedStates total Education level',
                     legend=False)
```

```python
USAEduLevel_3region = USAEduLevel.loc[USAEduLevel['country'].isin(
    ['United States','California', 'Florida', 'Texa'])]
USAEduLevel_3region
```

```python
USAEduLevel_3region_pivot = USAEduLevel_3region.pivot_table(values='Percentage%',
                                                    columns='country',
                                                    index='year')
USAEduLevel_3region_pivot
```

```python
USAEduLevel_3region_pivot.plot(title='Total education level in 3 states  of US')
plt.show()
```
